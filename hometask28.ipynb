## Let's get started 

hozirda men diamonds nomli tayyor methodi bilan ishlamoqchiman 

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

## let's create **scotter** ****plot**** with **seaborn** library

```python
sns.scatterplot(x="carat",y="depth",data=diamonds)
plt.show()
```

```python
diamonds_subset = diamonds.head()
sns.scatterplot(x="carat", y="depth", data=diamonds_subset, hue="cut")
plt.show()
```

bunda '**hue**' orqali graph da har bittasini ajratib olsak buladi, shuningdek **hue** bu yerda tayyor method hisoblanib aftomatiskiy uzi ajratib beradi *turli xil ranglar* bilan, shuning bilan birga oson bulishi uchun hozircha **head** funksiyasini ham ishlatildi

```python
sns.scatterplot(x="carat",y="depth",data=diamonds,hue="cut",hue_order=["Good","Fair","Very Good","Ideal","Premium"])
plt.show()
```

Bunda **Hue** parametrini oder buyicha taxlab chiqdim nu bu degani **Hue** METHOD ini uzimiz boshqarolishimizni bildiradi 

```python
hue_colors = {"Good": 'yellow',"Fair":"red","Very Good":"blue","Ideal":"green","Premium": "grey"}
sns.scatterplot(x="carat",y="depth",data=diamonds_subset,hue="cut",palette=hue_colors)
plt.show()
```

bu yerdan ham bilish mumkinki ranglarigacha biz murojat qilib bemolol **uzgartirolamiz**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x="carat",data=diamonds,hue="cut")
plt.show()
```

## head bilan kurilishi

```python
diamonds = sns.load_dataset('diamonds')
diamonds_subset = diamonds.head()
sns.countplot(x="carat", data=diamonds_subset, hue="cut")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
d = diamonds[(diamonds['cut'] == 'Good') | (diamonds['cut'] == 'Fair')]
sns.countplot(x="carat",data=d,hue="cut")
plt.show()
d
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = diamonds
# Assuming 'df' is your DataFrame and it has columns 'cut' and 'carat'
# Filter the DataFrame to include only 'Good' and 'Fair' cuts
filtered_df = df[df['cut'].isin(['Good', 'Fair'])]

# Now plot the filtered data
plt.figure(figsize=(10, 6))
sns.histplot(data=filtered_df, x='carat', hue='cut', multiple='stack')
plt.show()
```
