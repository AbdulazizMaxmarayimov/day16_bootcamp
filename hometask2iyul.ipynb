```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Data.csv', encoding='latin1')
df.head()
```

## Plot types - show each observation

```python
sns.stripplot(data=df, y="Country",x="xG Per Avg Match",jitter=True, hue="Country")
```

stripplot da ishlab kurish aniq bir graph larni olib

```python
sns.swarmplot(data=df, y="Country",x="xG Per Avg Match",hue="Country")
```

swarmplot da qullab kurish

```python
sns.boxplot(data=df, y="Country",x="xG Per Avg Match")
```

boxplot da qullab kurish abstak holatda


```python
sns.violinplot(data=df, y="Country",
x="xG Per Avg Match")
```

violinplot da qullab kurish

```python
sns.boxenplot(data=df, y="Country",
x="xG Per Avg Match")
```

boxenplot -> da kurib chqish graphni

```python
sns.barplot(data=df, y="Country",x="xG Per Avg Match",hue="Country")
```

barplot da kurib chiqish

```python
sns.pointplot(data=df, y="Country",x="xG Per Avg Match",hue="Country")
```

pointplot da kurib chiqish

```python
sns.countplot(data=df, y="Country",hue="Country")
```

countplot da faqat y ni hisoblab chiqish

## Plotting with regplot()

```python
sns.regplot(data=df, x='Matches_Played',y='On Target Per Avg Match', marker='+')
```

bunda hoz men regplot da urtadan chiziq utkazdim va bunda shuni kurish mumkinki ikki qismga ajratib uqilishga ancha oson qildi

```python
sns.residplot(data=df, x='Matches_Played',y='On Target Per Avg Match')
```

Seaborn supports through **residplot** function va bunda ham tepadagi bilan deyarli bir xir
\

```python
sns.regplot(data=df, x='Matches_Played',y='On Target Per Avg Match',order=2)
```

bunda order functioni orqali standar divison qilib teng ikkiga bulib chiqqan

```python
sns.regplot(data=df, x='Matches_Played',y='On Target Per Avg Match',x_jitter=.1,order=2)
```

x_jitter=.1 -> shu function qushilib ishlatildi

```python
import numpy as np
sns.regplot(data=df, x='Matches_Played',y='On Target Per Avg Match',x_estimator=np.mean,order=2)
```

bunda numpy va estimators functionlarini qullab highlight qilingan, taxminiy hisoblasa bu numby functioni qulayro bunda mean topilgan

```python
sns.regplot(data=df, x='Matches_Played',y='On Target Per Avg Match',x_bins=4)
```

x_bins maʼlumotlarni diskret joylarga boʻlish uchun ishlatilishi mumkin



```python
pd.crosstab( df["Goals"],df["Mins"],values=df["Matches_Played"],aggfunc="mean").round(0)
```

bunda hozir table yaratildi va bu table da goals va minuts larni oldim shunda biz kurishimiz mumkinki bunda nech minutda gol bulganini

```python
pivot_table = df.pivot_table(index='Goals', columns='Mins', values='Matches_Played', aggfunc='mean').round(0)
sns.heatmap(pivot_table, annot=True, fmt='g')
plt.show()
```
