## Interpreting boxplots

```python
fig, ax = plt.subplots()
ax.boxplot([economy_USA["year"].head(12)])
ax.set_xticklabels(["year"])
ax.set_ylabel("Height (cm)")
plt.show()
```

## Scatter plots

```python
fig, ax = plt.subplots()
ax.scatter(economy_USA["year"], economy_USA["high_school"])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(economy_USA["year"].head(6), economy_USA["high_school"].head(6), color = 'green')
ax.set_title('Employment-to-Population Ratio for USA', loc='left', fontsize=12, color = 'blue')  
ax.set_xlabel('Y E A R')
ax.set_ylabel('high_school (percentage %)')
plt.show()
```

bunda ichidagi har yillikni anniq kurish uchun va shuningdek label lar yaratish


```python
fig, ax = plt.subplots()
ax.scatter(economy_USA["year"], economy_USA["high_school"], c=economy_USA.index)
ax.set_xlabel('Y E A R')
ax.set_ylabel('high_school (percentage %)')
plt.show()
```

## hammasini bitta table ga joylashitrish

```python
plt.style.use("ggplot")
# ax.set_title('Employment-to-Population Ratio for USA', loc='left', fontsize=12, color = 'blue')  
fig, ax= plt.subplots(2,2)
ax[0,0].scatter(economy_USA["year"], economy_USA["high_school"], c=economy_USA.index)
ax[0,0].set_xlabel("PRCP", color="b")
ax[0,0].set_ylabel("TAVG", color="r")
ax[0,0].set_title("SEATTLE", color="y")

ax[1,0].bar(("2020"), economy_USA["high_school"].mean(), yerr=economy_USA["high_school"].std())
ax[1,0].bar(("2019"), economy_USA["high_school"].mean(), yerr=economy_USA["high_school"].std())
ax[1,0].set_ylabel("Height", color="b")
ax[1,0].set_xlabel("Y e a r", color="b")


ax[0,1].hist(economy_USA["year"].head(6), label=economy_USA['high_school'].head(6))
ax[1,1].boxplot([economy_USA["high_school"].head(12)])
ax[1,1].set_xticklabels(["year"])
ax[0][0].grid(True, color="b")
fig.set_size_inches([7,7])
plt.show()
```
