```python
import seaborn as sns
sns.displot(df['release_year'], kind='kde')
```

it generates a **histogram** but can also generate other plot types

```python
sns.displot(df['release_year'], kde=True, bins=10)
```

You can overlay a **KDE** plot on the *histogram and specify* the number of bins to use

```python
sns.displot(df['release_year'], kind='kde', rug=True, fill=True)
```

A **kde curve** and **rug plot** can be combined yani kde curve va rug plot combine bula oladi

```python
sns.displot(df['release_year'], kind='ecdf')
```

```python
styles = ['white', 'dark', 'whitegrid', 'darkgrid', 'ticks']

fig, axes = plt.subplots(1, len(styles), figsize=(20, 5))

for ax, style in zip(axes, styles):
    sns.set_style(style)
    sns.histplot(df['type'], ax=ax)
    ax.set_title(f"Style: {style}")

plt.show()
```

**Theme examples with sns.set_style()** misollar style buyicha

```python
sns.set_style('white')
sns.displot(df['type'])
sns.despine(left=True)
```

```python
palettes = ['deep','muted','pastel','bright','dark','colorblind']
for p in palettes:
    sns.set_palette(p)
    sns.displot(df['type'])
```

Seaborn uses the **set_palette**() function to define a palette -> bunda set palette buyicha uzgartirish
